# THEMIS-A ESA ion eflux, 2019-05-01

Single-probe look at the storm-time interval. Imports helpers from `esa_plotting` so the notebook stays thin.

## Setup

`autoreload` makes edits to `src/esa_plotting/` show up without restarting the kernel.

```python
%load_ext autoreload
%autoreload 2

from esa_plotting import load_esa, configure_eflux_panel, eflux_var, set_data_dir
from pyspedas import tplot, tplot_options, get_data

set_data_dir()
```

## Load data

```python
probe = "a"
trange = ["2019-05-01", "2019-05-02"]

load_esa(probe, trange)
v = eflux_var(probe)
v
```

## Inspect before plotting

Sanity check: how many time samples, how many energy bins.

```python
d = get_data(v)
print("times:", d.times.shape)
print("y    :", d.y.shape)
d.y[:3, :5]
```

## Plot

`configure_eflux_panel` applies the standard log-log-log + spedas colormap settings. Override the defaults by passing `zrange=...` or `yrange=...`.

```python
configure_eflux_panel(v, ytitle=f"ESA i+ th{probe}\nE [eV]")
tplot_options("title", f"THEMIS-{probe.upper()} ESA ion eflux {trange[0]}")
tplot(v, xsize=12, ysize=3)
```

- Storm onset visible around ~12:00 UT, broad energy spread up to 30 keV.
